```python
# Tristan Ursell
# G2 Decay Rate Constant Plot
# May 2024
#

import numpy as np
import matplotlib.pyplot as plt

# Generate values for ab
ab = np.arange(0, 2.0001, 0.0001)

# Compute function values
k_P = -np.log(np.abs(1 - ab))

# Plot function
plt.figure(figsize=(6, 5))
plt.plot(ab, k_P, linewidth=2, color=[0, 0.2, 0.7])
plt.axvline(x=1, color='k', linestyle='--')

# Labels and formatting
plt.xlabel('$(a + b)$')
plt.ylabel('$k_P$')
plt.box(True)
plt.xlim([0, 2])
plt.ylim([0, 10])
plt.savefig(f'G2_decay_rate_constant.svg', format='svg')
plt.show()
```
