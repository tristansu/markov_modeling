```python
# Tristan Ursell
# Single-Player Betting Game
# May 2024
#

import numpy as np
import matplotlib.pyplot as plt

# Values of p to sample (high resolution)
p = np.arange(0.5, 1.0001, 0.0001)

# Starting dollar amounts to sample
D = np.array([1, 2, 10, 100])

# Winning probabilities to sample
wstar = np.array([0.25, 0.5, 0.75, 0.95, 0.99])

# Figure 1: Bankruptcy Probabilities
plt.figure()
cmap1 = plt.get_cmap('jet', len(D))
for i, d in enumerate(D):
    bD = (1 / p - 1) ** d
    plt.plot(p, bD, linewidth=2, color=cmap1(i))

plt.plot([0, 0.5], [1, 1], 'k-', linewidth=2)
plt.xlim([0, 1])
plt.ylim([-0.025, 1.025])
plt.box(True)
plt.xlabel('Heads probability (p)')
plt.ylabel('Bankruptcy Probability (b_D)')
plt.legend([f'D = {d}' for d in D], loc='upper right')
plt.savefig(f'single_player_betting_bankruptcy_prob.svg', format='svg')
plt.show()

# Figure 2: Betting Tolerances
plt.figure()
cmap2 = plt.get_cmap('jet', len(wstar))
for i, w in enumerate(wstar):
    nstar = 1 + np.log(1 - w) / np.log(1 / p - 1)
    plt.semilogy(p[1:], nstar[1:], linewidth=2, color=cmap2(i))

plt.semilogy([0.5, 0.5], [1, 100], 'k--')
plt.xlim([0, 1])
plt.ylim([1, 100])
plt.box(True)
plt.xlabel('Heads probability (p)')
plt.ylabel('D to win @ w*')
plt.legend([f'w* = {w}' for w in wstar], loc='upper right')
plt.savefig(f'single_player_betting_critical_vals.svg', format='svg')
plt.show()
```
